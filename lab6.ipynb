Pull data from Piedra 

```r
piedra = readNWISdata(sites='09349800',service='dv',startDate="2017-10-01",endDate="2018-09-30")
piedra2 = readNWISdata(sites='09349800',service='dv',startDate="2016-10-01",endDate="2017-09-30")
```

```r
plot(piedra$dateTime,piedra2$X_00060_00003,type='l',col='red')
lines(piedra$dateTime,piedra$X_00060_00003,type='l')
p_cdf = ecdf(piedra$X_00060_00003)
p2_cdf = ecdf(piedra2$X_00060_00003)
plot(p_cdf,main="Piedra River, Gage (09349800)",xlab = 'Water Year Day',ylab="CDF")
plot(p2_cdf,add=T,col='red')
legend(1, 1, legend=c("WY2018", "WY2017"),
       col=c("black", "red"), lty=1, cex=0.8)


```
